### Preparing a plot for a scientific figure

**Size and font size**
- Page width is often around 7 inches / 180 mm.
- Two plots next to each other should each be around 3 inches wide.
- Use a small but readable font size, e.g. 8–10 pt.
- SVG text should stay editable in Inkscape.

**Improve visuals**
- Remove unnecessary borders, grid lines, and labels.
- Draw attention to the important elements.
- Avoid 3D effects and unnecessary colors.


```python
# Use a smaller font size for figure panels
plt.rcParams["font.size"] = 8
```

## Bonus: Colormaps

Use colormaps deliberately:
- **Qualitative**: categories without numerical order
- **Sequential**: low-to-high numerical values
- **Diverging**: values around a meaningful midpoint


```python
# Qualitative colormaps
sns.color_palette("tab10")
```

```python
sns.color_palette("pastel")
```

```python
# Sequential colormap
sns.color_palette("viridis", as_cmap=True)
```

```python
# Diverging colormap
sns.color_palette("coolwarm", as_cmap=True)
```

```python
# Heatmap example without relying on an online dataset download
example = pd.DataFrame(
    np.arange(1, 17).reshape(4, 4),
    index=["A", "B", "C", "D"],
    columns=["W", "X", "Y", "Z"],
)

sns.heatmap(example, cmap="viridis", annot=True)
plt.show()
```
